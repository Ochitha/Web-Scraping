```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the website
url = "https://www.boc.lk/rates-tariff#rupee-deposit-rates"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table element by its class name
table = soup.find(class_='ck-table-resized')

# Extract the table headers
headers = [header.text.strip() for header in table.find_all('th')]

# Extract the table rows
rows = []
for row in table.find_all('tr'):
    row_data = [data.text.strip() for data in row.find_all('td')]
    rows.append(row_data)

rows = rows[1:]
# Create a DataFrame from the extracted data
df = pd.DataFrame(rows, columns=headers)

# Print the DataFrame
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the website
url = "https://www.nationstrust.com/deposit-rates"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table element by its class name
table = soup.find(class_='table table-bordered')

# Extract the table headers
headers = [header.text.strip() for header in table.find_all('th')]

# Extract the table rows
rows = []
for row in table.find_all('tr'):
    row_data = [data.text.strip() for data in row.find_all('td')]
    rows.append(row_data)

# Extract the data rows
data_rows = rows[1:]

# Create a DataFrame from the data rows
df = pd.DataFrame(data_rows)

# Transpose the DataFrame
transposed_df = df.transpose()

# Assign the second row as the new column names
transposed_df.columns = transposed_df.iloc[1]

# Drop the second row from the DataFrame
transposed_df = transposed_df.drop(1)



# Reset the index of the DataFrame
transposed_df = transposed_df.reset_index(drop=True)


# Print the transposed DataFrame
transposed_df
```
