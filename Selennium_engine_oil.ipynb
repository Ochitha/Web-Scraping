#Caltex Engine oil

```python
#beautifulsoup
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the website
url = "https://greasemonkey.lk/product-category/engine-oil-lubricants/caltex/"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the product names and prices in nested tags using the main tag
oil_name = soup.find_all('h2', class_='woocommerce-loop-product__title') # Find all the product names and prices in nested tags using the main tag
#land_perches = soup.find('ul', class_='list--3NxGO').find_all('div', attrs={'class': None})
oil_price = soup.find_all('span', class_='price')

# Create empty lists to store the names and prices
oil_label = []
prices = []

# Extract the text content from the HTML tags and append to the lists
for name in oil_name:
    oil_label.append(name.text.strip())
for price in oil_price:
    prices.append(price.text.strip())

# Create a pandas DataFrame to store the names and prices
df = pd.DataFrame({'Name': oil_label,'Price': prices})

# Print the DataFrame
print(df)

# Save the DataFrame to a CSV file
#df.to_csv("medicine_data.csv", index=False)
```

```python
#selenium with loop
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests

url = "https://greasemonkey.lk/product-category/engine-oil-lubricants/caltex/"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Create empty lists to store the names and prices
oil_label = []
prices = []

#loop
for page in range(1,6, 1):

    page_url = "https://greasemonkey.lk/product-category/engine-oil-lubricants/caltex/page/" + str(page)
    driver = webdriver.Chrome()
    driver.get(page_url)
    land_location = soup.find_all('div', class_='description--2-ez3')
    # Find all the product names and prices in nested tags using the main tag
    oil_name = soup.find_all('h2', class_='woocommerce-loop-product__title') # Find all the product names and prices in nested tags using the main tag
    #land_perches = soup.find('ul', class_='list--3NxGO').find_all('div', attrs={'class': None})
    oil_price = soup.find_all('span', class_='price')

    for name in oil_name:
        oil_label.append(name.text.strip())
    for price in oil_price:
        prices.append(price.text.strip())


# Create a pandas DataFrame to store the names and prices
df = pd.DataFrame({'Name': oil_label,'Price': prices})


# Print the DataFrame
df.to_csv("oil_data.csv", index=False)
driver.close()
```
