```python
#one page selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date
import time

# Initialize a new Chrome browser using Selenium
driver = webdriver.Chrome()

# Navigate to the website
driver.get('https://www.onlinekade.lk/product-category/stationery/')

# Wait for the page to load
time.sleep(10)

# Scroll down to the end of the page to load all products
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Extract the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all the product names and prices using their HTML tags and classes
product_names = soup.find_all('h3', class_='product-title')
product_prices = soup.find_all('bdi')

# Create empty lists to store the names and prices
names = []
prices = []

# Extract the text content from the HTML tags and append to the lists
for name in product_names:
    names.append(name.text)
for price in product_prices:
    prices.append(price.text)

# Create a pandas DataFrame to store the names, prices and today's date
today = date.today()
prices_new=[value for value in prices if value != "Rs.0.00"]
df = pd.DataFrame({'Name': names, 'Price': prices_new, 'Date': [today]*len(names)})

# Print the DataFrame
df.to_csv("stationery2.csv", index = False)

# Close the browser
driver.quit()
```
